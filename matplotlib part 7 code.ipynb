```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\rjhcl\Desktop\07-ScatterPlots\data.csv')
```

```python
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']
```

```python
plt.scatter(view_count , likes , c = ratio , edgecolor = 'black')
cbar = plt.colorbar()
cbar.set_label('ratio of likes\dislikes')
plt.xscale('log')
plt.yscale('log')
plt.title('Youtube videos analysis')
plt.xlabel('Viewers')
plt.ylabel('likes')
```
